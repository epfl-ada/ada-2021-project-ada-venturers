```python
import pandas as pd
import numpy as np
data_pickles = ['../data2015.pkl', '../data2016.pkl','../data2017.pkl', '../data2018.pkl','../data2019.pkl', '../data2020.pkl']
dfs = [pd.read_pickle(f) for f in data_pickles]
df = pd.concat(dfs)
df
```

```python
df_time = df.groupby(pd.Grouper(freq='W', key='date'), as_index=True).sum()
df_time = df_time.rename(columns={'numOccurrences':'val'})
df_time
```

```python
def add_annotation(fig, df_time, date, text):   
    fig.add_annotation(
            x=date,
            y=df_time.loc[date].val,
            text=text,
            showarrow=True,
            align="center",
            arrowhead=2,
            arrowsize=1,
            arrowwidth=2,
            borderwidth=2,
            borderpad=4,
            opacity=0.8,
            bgcolor="#EFEFEF"
            )
```

```python
import plotly.express as px
import plotly.graph_objects as go
import datetime

fig = go.Figure()


fig.add_trace(go.Scatter(x=df_time.index, y=df_time.val))
fig.update_xaxes(rangeslider_visible=True)

add_annotation(fig, df_time, datetime.datetime(2017, 6, 4), "Trump leaves the Paris Agreement")
add_annotation(fig, df_time, datetime.datetime(2015, 12, 6), "Trump gets elected")

fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)

fig.show()
```
